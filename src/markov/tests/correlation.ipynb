```python
import json
import numpy as np
from networkx.readwrite import json_graph
from fight import Fight
```

```python
PATH_G = '/Users/huibmeulenbelt/PycharmProjects/ufc/scripts/markov/tests/12120.json'
G = json_graph.adjacency_graph(json.load(open(PATH_G)))
```

```python
n_rounds = 3
fight = Fight(G=G, n_rounds=n_rounds)
```

```python
Sigma = fight.Sigma
```

```python
# check for asymmetries
Sigma[0]
```

```python
Sigma[0].T
```

```python
delta = np.abs(Sigma - np.transpose(Sigma, axes=(0, 2, 1)))
bool(np.allclose(delta, 0))
```

```python
sigma_sym = 0.5 * (Sigma * np.transpose(Sigma, axes=(0, 2, 1)))      # enforce symmetry
```

```python
std = np.sqrt(np.diagonal(sigma_sym, axis1=1, axis2=2))
inv_std = 1.0 / std
```

```python
corr = sigma_sym * inv_std[:, :, None] * inv_std[:, None, :]
corr[:, np.arange(fight.r), np.arange(fight.r)] = 1.0                # diagonal is always 1
```

Some states are linearly dependent on each other so will have a correlation of 1.
The correlation should be in the interval $[-1, 1]$.

```python
bool(np.all(abs(corr) <= 1))
```
