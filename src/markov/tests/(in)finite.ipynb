```python
import json
import numpy as np
from networkx.readwrite import json_graph
```

```python
from fight import Fight
from database.database import get_markov_state
```

```python
PATH_G = '/Users/huibmeulenbelt/PycharmProjects/ufc/scripts/markov/tests/12120.json'
G = json_graph.adjacency_graph(json.load(open(PATH_G)))
```

```python
n_rounds = 3
fight = Fight(G=G, n_rounds=n_rounds)
```

```python
idx = fight.r_nodes_idx
inv_idx = dict(map(reversed, idx.items()))
```

```python
N_n = fight.N
N_inf = np.linalg.matrix_power(fight.I - fight.Q, -1)
```

```python
N_n[:,0]
```

```python
N_inf[:,0]
```

```python
delta = abs(N_inf[:,0] - N_n[:, 0])
```

```python
delta_avg = np.mean(delta, axis=0)
```

```python
for i, (k, v) in enumerate(inv_idx.items()):
    abbr, fighter = get_markov_state(markov_id=v)
    print(f'{abbr}_{fighter}:\t{round(delta_avg[i], 2)}')
```
